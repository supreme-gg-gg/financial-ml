```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('../data/GOOG.csv')
# Preview the first five entries
print(data.head())

# Plot the adjusted close over time
plt.plot(data['Date'], data['Adj Close'])
plt.xlabel('Date')
plt.ylabel('Adjusted Close')
plt.title('Adjusted Close Over Time')
plt.show()
```
